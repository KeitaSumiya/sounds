```python
import numpy as np
import wave
import matplotlib.pyplot as plt
```

```python
output_wave = []
freq=2
num_of_point=0.001
for n in np.arange(-1,1,num_of_point):
    s = np.sin(2.0*np.pi*freq*n)
    output_wave.append(s)
```

```python
plt.plot(output_wave,'o')
```

```python
s_output_wave = []
freq = 2
sample_freq = 10 #New!
num_of_point=0.001*sample_freq
for n in np.arange(-1,1,num_of_point):
    s = np.sin(2.0*np.pi*freq*n)
    s_output_wave.append(s)
```
